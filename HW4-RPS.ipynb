### mybot.py code

The algorithm uses a combination of CTW and LZ. The sequential probability obtained from both algorithms would then be combined to to generate one set of probabiliies, corresponding to R,P and S. The probability values generated are also perturbed using uniform normal distribution, with zero mean and decreasing variance (variance decreases after each iterations, so that at first variance us high, giving weights to random probability than that calculated using LZ or CTW).

This idea could work, as LZ has the lowest regret for stationary ergodic processes, and CTW has better convergence properties than LZ for certain cases. Since we are using a combination of both, when both algorithms strongly favor a move, that is played. Otherwise, the move will be played by the algorithm with more confidence.
